```python
df.describe()
#mostra dados como quantidade, média, desvio padrão ('std')
#os dados em 25%, 50% e 75% indicam os quartis que...
#...são a maneira pela qual os dados estão distribuidos...
#...em ordem crescente: o dado em 25% indica o primeiro quarto; o dado em 50% apresenta...
#...o dado que está exatamente no meio; e o dado em 75% apresenta o número que consta em...
#...3/4 da sequência crescente
#Isso indica, por exemplo que metade dos pokemons tem vida em até 65 pontos!
#Ou que 1/4 dos pokemons tem pontos de vida iguais ou menores que 50.  
#Através desses dados podemos ver também que entre o primeiro quartil (25%)
#e o próximo (50%), temos outro quarto que tem entre 50 e 65 pontos de vida
```
